```python
import pandas as pd

race_data = pd.read_csv('pres_prediction_data.csv')
race_data
```

```python
import statsmodels.api as sm

Y = race_data['share']
X = race_data[['%_white_alone', '%_no_bachelors', 'median_income']]

model = sm.OLS(Y, X).fit()
print(model.summary())
```
