```python
ind = df.iloc[0, :]
ind = ind[4:18]
ind = pd.to_numeric(ind, errors = 'coerce')

count = 0;
c = [-1000]

for i in range(1, len(df['area_name'])):
    row = df.iloc[i, :]
    row = row[4:18]
    row = pd.to_numeric(row, errors = 'coerce')

    df2 = pd.DataFrame(data=ind)
    df2['values'] = row
    df2.rename(columns={0: 'index'}, inplace=True)
    
    X = df2['index']
    Y = df2['values']
    X = sm.add_constant(df2['index'])
    
    model = sm.OLS(Y, X).fit()
    coef = model.params[1]
    
    c.append(coef)
    
    count = count + 1;

df = df.assign(e=pd.Series(c).values)
df.head()

#df.to_csv(path_or_buf= 'trend.csv', index= False)
```
